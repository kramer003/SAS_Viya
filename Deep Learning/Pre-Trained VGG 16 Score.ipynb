### Helper Function

```python
def image_json(image_path):
    with open(image_path, "rb") as image_file:
        image_string = base64.b64encode(image_file.read()).decode()
        image_string_dict = {'val':image_string}
    return image_string_dict
```

```python
import requests
from requests.auth import HTTPBasicAuth

data = image_json(PATH)

# Post request body with basic auth
auth=('andrewkramer', 'ef33e062-89ce-44a2-9c96-3fb93f75256e')
r = requests.post('http://10.10.9.50/production/models/VGGtf/predict', json = data, auth = auth)

# Print response
print(r.text)
```
